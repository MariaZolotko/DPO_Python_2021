```python
import math as math
```

```python
def f(x):
    return math.sin(x/5.0)*math.exp(x/10.0)+5.0*math.exp(-x/2.0)
x1=1.0
x2=15.0
x3=8.0
x4=4.0
x5=10.0
y1=f(x1)
y2=f(x2)
y3=f(x3)
y4=f(x4)
y5=f(x5)
print("y(1):",y1)
print("y(15):",y2)
print("y(8):",y3)
print("y(4):",y4)
print("y(10):",y5)
```

```python
import numpy as np
```

```python
a = np.array([[1.0,x1], [1.0,x2]])
b = np.array([y1, y2])
x = np.linalg.solve(a, b)
print(x)
```

```python
a = np.array([[x1**3,x1**2,x1,1.0], [x3**3,x3**2,x3,1.0],[x4**3,x4**2,x4,1.0],[x2**3,x2**2,x2,1.0]])
b = np.array([y1,y3,y4,y2])
x = np.linalg.solve(a, b)
y = np.array(x[[3,2,1,0]])

xstr=str(y.tolist()).strip('[]') 

print(xstr)
```
